```python
countries_gdphist = df.loc[:,'1960':'2018'].transpose()
```

```python
gdp_ROC = {} #dictionary of dictionaries with gdp history

for country, gdphist in countries_gdphist.iteritems(): #start row iteration
    
    cur = None
    cur_i = None
    prev = None
    prev_i = None
    year_ROC = {}
    
    for i, (year, gdp) in enumerate(gdphist.iteritems()): #start column iteration
        if pd.notnull(gdp): # save item that is not null
            cur = gdp
            cur_i = i
            
        if pd.isnull(prev): # stablish 1st prev: once per column iteration
            prev = cur
            prev_i = cur_i
        
        if cur_i != prev_i: #check if operation can be done. Cur_i == prev_i only when loop
            #has not found a new pair
            ROC = (cur - prev) / (cur_i - prev_i) # perform operation for Rate Of Change
    
            
            year_ROC[year] = ROC/prev*100 #normalize rate of change
            
            print(ROC, prev)
            
            prev = cur # set prev to cur for next loop
            prev_i = cur_i #set prev to cur for next loop
            
    gdp_ROC[country] = year_ROC
```

```python
gdp_ROC_df = pd.DataFrame(gdp_ROC).transpose() #create dataframe from dictionary
```

```python
gdp_ROC_df.sort_index(inplace=True, axis=1) #the dictionary produced a dataframe that was not ordered
```
