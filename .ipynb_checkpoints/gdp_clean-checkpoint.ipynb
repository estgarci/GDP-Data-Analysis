# Data Stratification

## GDP increase

```python
import numpy as np
from collections import OrderedDict

def rate_normalized(dataFrame):
    '''
    calculates rate of change and returns a normalized ROC
    value based on the previous value, operates on columns
    ''' 
    
    rate_of_change = OrderedDict() #dictionary of dictionaries with gdp history
    for country, gdpHist in dataFrame.iterrows(): #start row iteration

      
        cur = None
        cur_i = None
        prev = None
        prev_i = None
        year_ROC = OrderedDict()
        
        for i, (year, gdp) in enumerate(gdpHist.iteritems()): #start column iteration
            
            if pd.notnull(gdp): # save item that is not null
                cur = gdp
                cur_i = i
                
            if pd.isnull(prev): # stablish 1st prev: once per column iteration
                prev = cur
                prev_i = cur_i

            if cur_i != prev_i: #check if operation can be done. Cur_i == prev_i only when loop
                #has not found a new pair
                ROC = (cur - prev) / (cur_i - prev_i) # perform operation for Rate Of Change
                
                year_ROC[year] = ROC/prev*100 #normalize rate of change
                
                prev = cur # set prev to cur for next loop
                prev_i = cur_i #set prev to cur for next loop

        rate_of_change[country] = year_ROC
        
    return rate_of_change
```

### US_presidents

```python
from bs4 import BeautifulSoup
import urllib
import re
from datetime import datetime, timedelta


#downloading united states presidents from wikipedia
us_p_ws = r'https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States'
us_p_html = urllib.request.urlopen(us_p_ws)
us_p_soup = BeautifulSoup(us_p_html, 'lxml')

#extracting table
table = us_p_soup.find_all('table')[1]
            
#extracting information from table
presidentNum = 0
presidents = {}

for tr in table.find_all('tr'):
    tds = tr.find_all('td')[:-1]
    
    if len(tds) > 5:
        presidentNum +=1
        
        #cleaning time
        t_start = re.sub( r'([A-Z])\w+', tds[1].text[:3], tds[1].text[:-1]).split('')[0]
        t_start = re.sub(r'\[.\]$', '', t_start)
        t_start = datetime.strptime(t_start, "%b %d, %Y").date()

        t_end = re.sub( r'([A-Z])\w+', tds[1].text[:3], tds[1].text[:-1]).split('')[1]
        t_end = re.sub(r'\[.\]$', '', t_end)
        
        
        try:
            t_end = datetime.strptime(t_end, "%b %d, %Y").date()
        except:
            t_end = datetime(2020, 6, 30)
        
        #adding extracted nformation to dictionary
        presidents[ re.sub(r'\[.\]$','',tds[3].text[:-1]) ] = {
            
            'President number': presidentNum,
            'Term Start': t_start,
            'Term End': t_end,
            'Political party': re.sub(r'\[.\]$', '', tds[5].text[:-1])     
        }
```

```python
US_presidents = pd.DataFrame(presidents).T

US_presidents[:5]
```

```python
def date_range(kind, start, end):
    '''
    Returns a list of dates
    
    [str] kind: specified time delta: month, day, or hour 
    [date] start: start date
    [date] end: end date
    '''
    
    l_range = []

    if kind == 'year':
        start_m = start.year
        end_m = end.year
        delta = timedelta(years=1)
        difference = end_m - start_m

    elif kind == 'month':
        start_m = start.month
        end_m = end.month
        delta =timedelta(months=1)
        difference = end_m - start_m

    elif kind == 'day':
        delta = timedelta(days=1)
        difference = end - start
        difference = difference.days
        
    else:
        return 'ERROR: Enter correct format for \'kind\': day, month, year'
    
    i = 0
    date = start
    while i <= difference:
        l_range.append(date)
        date = date + delta
        i += 1
        
    return l_range
```
